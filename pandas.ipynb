```python
# @title Folium: librería para mapas

import folium
```

```python
mapa = folium.Map(location=[-16.411551448369803, -71.5235309425812], zoom_start=50)
folium.Marker([-16.411551448369803, -71.5235309425812], popup="CERTUS", tooltip="Clic Aquí!", icon = folium.Icon(color="blue")).add_to(mapa)
mapa
```

```python
mapa_regiones = pd.read_excel("Sesión 11 - Datos Matrcula.xlsx", sheet_name="Hoja06")
mapa_regiones.head()
```

```python
df_dptos = mapa_regiones[["REGION", "Latitud", "Longitud"]]
df_dptos.head()
```

```python
for index, ubicaciones in df_dptos.iterrows():
  print(ubicaciones)
```

```python
mapa_peru = folium.Map(location=[df_dptos.Latitud.mean(), df_dptos.Longitud.mean()], zoom_start=5)
for index, ubicaciones in df_dptos.iterrows():
  folium.Marker([ubicaciones["Latitud"], ubicaciones["Longitud"]], popup=ubicaciones["REGION"]).add_to(mapa_peru)
mapa_peru
```
